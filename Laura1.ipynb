# Задание 8

Для спецификации log-log построим корреляционную матрицу:

```python
mod8_1 = smf.ols("np.log(body_mass_g) ~ np.log(bill_length_mm)+np.log(bill_depth_mm)+np.log(flipper_length_mm)+species+island-1", data=data).fit()
mod8_1.cov_params()
```

# Задание 9

Проведём тест Бройша-Пагана на гетероскедастичность для log-log модели:

```python
mod9_1 = smf.ols("np.log(body_mass_g) ~ np.log(bill_length_mm)+np.log(bill_depth_mm)+np.log(flipper_length_mm)+species+island-1", data=data).fit()
sm.stats.diagnostic.het_breuschpagan(mod9_1.resid, mod9_1.model.exog)
```

Второе значение это p-value. Гипотеза о гомоскедастичности отвергается на уровне значимости 5% и принимается на уровне значимости 1%.
